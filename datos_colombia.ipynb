```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

# Assuming the file is an Excel file, load it into a pandas DataFrame
file_path = '/content/drive/MyDrive/Colombia/Datos/datos_2006_2018.xlsx'
try:
    df_2006_2018 = pd.read_excel(file_path)
    print(f"DataFrame from '{file_path}' loaded successfully.")
    display(df_2006_2018.head())
except FileNotFoundError:
    print(f"Error: File not found at '{file_path}'. Please check the path.")
except Exception as e:
    print(f"Error loading {file_path}: {e}")
```

```python
# Assuming you have two dataframes named df_2006_2018 and another dataframe
# Let's assume the second dataframe is named df_new and is loaded from a different source
# If you have loaded the second dataframe using the previous code,
# you can access it from the `dfs` dictionary, e.g., df_new = dfs['your_second_file_name.csv']

# For demonstration purposes, let's create a dummy df_new
# Replace this with the actual loading of your second dataframe
data_new = {'CÓD. \nDEP.': [1, 2], 'DEPARTAMENTO': ['Test1', 'Test2'],
            'CÓD. MUN.': [101, 102], 'MUNICIPIO': ['City1', 'City2'],
            'GRUPO \nDE CULTIVO': ['Fruits', 'Vegetables'],
            'SUBGRUPO \nDE CULTIVO': ['Apple', 'Carrot'],
            'CULTIVO': ['Apple', 'Carrot'],
            'DESAGREGACIÓN REGIONAL Y/O SISTEMA PRODUCTIVO': ['Apple', 'Carrot'],
            'AÑO': [2019, 2020], 'PERIODO': ['2019A', '2020B'],
            'Área Sembrada\n(ha)': [10, 20], 'Área Cosechada\n(ha)': [8, 18],
            'Producción\n(t)': [50, 100], 'Rendimiento\n(t/ha)': [5, 5.5],
            'ESTADO FISICO PRODUCCION': ['Fresh', 'Fresh'],
            'NOMBRE \nCIENTIFICO': ['Malus domestica', 'Daucus carota'],
            'CICLO DE CULTIVO': ['Permanent', 'Transitory'], 'Unnamed: 17': [None, None]}
df_new = pd.DataFrame(data_new)


# Concatenate the two dataframes
# Assuming both dataframes have the same columns and order
# If not, you might need to align columns before concatenating
try:
    df_combined = pd.concat([df_2006_2018, df_new], ignore_index=True)
    print("DataFrames concatenated successfully.")
except Exception as e:
    print(f"Error concatenating dataframes: {e}")


# Define the output file path
output_file_path = '/content/drive/MyDrive/Colombia/Datos/datos_2006_2024.xlsx'

# Save the combined dataframe to an Excel file
try:
    df_combined.to_excel(output_file_path, index=False)
    print(f"Combined DataFrame saved to '{output_file_path}' successfully.")
except Exception as e:
    print(f"Error saving combined dataframe to Excel: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming the dataframe is named df_combined as in the previous step

# 1. Quantity of rows and columns
df_shape = df_combined.shape
print("DataFrame Shape:")
print(df_shape)
print("\n")

# 2. Quantity of missing data per column
missing_values = df_combined.isnull().sum()
print("Missing values per column:")
print(missing_values)
print("\n")

# 3. Quantity of unique crops
# Assuming the column name for crops is 'CULTIVO'
cultivo_col = 'CULTIVO'
if cultivo_col in df_combined.columns:
    num_unique_cultivos = df_combined[cultivo_col].nunique()
    print(f"Number of unique crops: {num_unique_cultivos}")
else:
    num_unique_cultivos = "Column 'CULTIVO' not found."
    print(num_unique_cultivos)
print("\n")

# 4. Data for the 10 crops with the largest planted area
# Assuming the column name for planted area is 'Área Sembrada\n(ha)'
area_sembrada_col = 'Área Sembrada\n(ha)'
if cultivo_col in df_combined.columns and area_sembrada_col in df_combined.columns:
    top_10_cultivos_area = df_combined.groupby(cultivo_col)[area_sembrada_col].sum().nlargest(10)
else:
    top_10_cultivos_area = "Columns required for top 10 crops by planted area not found."


# Define the output file path
output_file_path = '/content/drive/MyDrive/Colombia/Datos/datos_2006_2024.xlsx'

# Save the combined dataframe and analysis results to different sheets in the same Excel file
try:
    with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
        # Save the combined dataframe
        df_combined.to_excel(writer, sheet_name='Combined Data', index=False)

        # Save the summary statistics
        summary_data = {
            'Metric': ['DataFrame Shape', 'Number of Unique Crops'],
            'Value': [str(df_shape), str(num_unique_cultivos)]
        }
        summary_df = pd.DataFrame(summary_data)
        summary_df.to_excel(writer, sheet_name='Summary Statistics', index=False)

        # Save the missing values count
        if isinstance(missing_values, pd.Series):
             missing_values.to_excel(writer, sheet_name='Missing Values', header=['Missing Count'])
        else:
             # Handle the case where missing_values is a string (column not found)
             pd.DataFrame({'Error': [missing_values]}).to_excel(writer, sheet_name='Missing Values', index=False)


        # Save the top 10 crops by planted area
        if isinstance(top_10_cultivos_area, pd.Series):
            top_10_cultivos_area.to_excel(writer, sheet_name='Top 10 Crops by Area', header=['Total Planted Area (ha)'])
        else:
            # Handle the case where top_10_cultivos_area is a string (columns not found)
            pd.DataFrame({'Error': [top_10_cultivos_area]}).to_excel(writer, sheet_name='Top 10 Crops by Area', index=False)


    print(f"Combined DataFrame and analysis results saved to '{output_file_path}' successfully.")
except Exception as e:
    print(f"Error saving data to Excel: {e}")


# Display the bar plot of the 10 crops with the largest planted area
if isinstance(top_10_cultivos_area, pd.Series):
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_10_cultivos_area.index, y=top_10_cultivos_area.values, palette='viridis')
    plt.title('Top 10 Crops by Planted Area')
    plt.xlabel('Crop')
    plt.ylabel('Planted Area (ha)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
